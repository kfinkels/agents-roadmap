## ðŸ›¤ï¸ Which Agent should you build?

> [!TIP]
> **Use ReAct if:**
> - You are building a **Chatbot** where the user might change their mind mid-task.
> - The tool outputs are **unpredictable** (e.g., searching the live web).
> - The agent needs to ask **clarifying questions** ("Which customer do you mean?").

> [!IMPORTANT]
> **Use Plan-and-Execute if:**
> - You are building an **Automation Pipeline** (e.g., weekly financial reporting).
> - You need an **Audit Trail** (someone must approve the plan before it runs).
> - **Cost and token efficiency** are priorities for long, multi-step tasks.

## ðŸš€ The Hybrid Approach

In production, the best agents are often **Hybrid**.

* The **Planner** creates a high-level strategy (Step 1: Audit, Step 2: Refund).
* The **ReAct Executor** is used inside each step to handle any small surprises or errors that occur during execution.

This gives you the reliability of a plan with the flexibility of a reasoning loop.
